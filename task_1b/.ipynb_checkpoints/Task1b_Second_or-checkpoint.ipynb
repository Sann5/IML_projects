## Libraries import

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

## Read input data 

```python
path = ""                            ## Select the path to the data file
filename = "train.csv"                 
train_data = pd.read_csv(path+filename)
train_data
```

## Create new features

```python
# Quadratic features
train_data["phi6"] = train_data["x1"]**2
train_data["phi7"] = train_data["x2"]**2
train_data["phi8"] = train_data["x3"]**2
train_data["phi9"] = train_data["x4"]**2
train_data["phi10"] = train_data["x5"]**2

# Exponential features
train_data["phi11"] = np.exp(train_data["x1"])
train_data["phi12"] = np.exp(train_data["x2"])
train_data["phi13"] = np.exp(train_data["x3"])
train_data["phi14"] = np.exp(train_data["x4"])
train_data["phi15"] = np.exp(train_data["x5"])

# Cosine features
train_data["phi16"] = np.cos(train_data["x1"])
train_data["phi17"] = np.cos(train_data["x2"])
train_data["phi18"] = np.cos(train_data["x3"])
train_data["phi19"] = np.cos(train_data["x4"])
train_data["phi20"] = np.cos(train_data["x5"])
# Constan features

train_data["phi21"] = 1
```

```python
train_data
```

## Train a linear classifier and determine the weights of the features

```python
X = train_data.iloc[:, 2:23].values          ## Select the featuers
y = train_data.iloc[:,1].values              ## Select the class labels
reg = LinearRegression().fit(X, y)
coeficients = reg.coef_
pd.DataFrame(coeficients).to_csv("output.csv", index = False, header = False)
```

```python
reg.coef_
```

```python
from sklearn.linear_model import Lasso
reg = Lasso(alpha=0.0001, fit_intercept=False, max_iter=10000).fit(X,y)

coeficients = reg.coef_
pd.DataFrame(coeficients).to_csv("output.csv", index = False, header = False)
```

```python
lasso = Lasso()
              
alphas = [0.00000001,0.0001 0.001, 1, 10, 100, 200]                          ## Specify the hyperparameters


def RMSE_function(y_test, y_pred):
    """Function that returns the RSME 
    given the predicted class labels and
    the actual class labels"""
    RSME = np.sqrt(mean_squared_error(y_test, y_pred))
    return RSME

RMSE = make_scorer(RMSE_function)    ## Specify the scoring scheme
parameters = [{'alpha': alphas}]                         ## Put hyperparameters into a tupple
n_folds = 5                               
```

```python
clf = GridSearchCV(estimator = lasso,           ## Create a ridge regression classifier 
                   param_grid = parameters, 
                   cv = n_folds,                ## Perform a cross-validation of the parameters n_folds
                   scoring = RMSE,              ## Output the RSME score of our results
                   refit = False)

clf.fit(X, y)                                                      ## Fit the data
scores = pd.DataFrame(clf.cv_results_['mean_test_score']) ## Save the results to a dataframe
scores.to_csv('output.csv', index = False, header = False)             ## Save the results dataframe to a csv
print(scores) 
```
